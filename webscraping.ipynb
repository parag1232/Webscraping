```python
import pandas as pd
```

```python
#Web scraping
from selenium import webdriver
path="C:/Users/parag/Downloads/chromedriver.exe"
driver=webdriver.Chrome(path)
driver.get("https://www.topuniversities.com/university-rankings/world-university-rankings/2018")
```

```python
a=driver.find_element_by_class_name("dataTable")#run this  line for every page
b=a.find_elements_by_class_name("title")
```

```python
for i in b:
    uni.append(i.text)
```

```python
len(uni)
```

```python
uni=pd.DataFrame(uni,columns=["University name"])
```

```python
rank=[]
```

```python
for i,j in enumerate(uni["University name"]):
    rank.append(i+1)
```

```python
uni["Rank as 2018"]=rank
```

```python
uni.tail()
```

```python
uni.to_csv("top100universities.csv",header=True,index=False,encoding="UTF-8")
```
