```python
from googlemaps import places
import urllib
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df=pd.read_csv("unicourse.csv")
df
```

```python
df["Latitude"]=df.Latitude.astype(str)
df["Longitude"]=df.Longitude.astype(str)
df["latlong"]=df.Latitude+","+df.Longitude
```

```python
df.head()
```

```python
#creating lists to take data from api
name=[]
lat=[]
lng=[]
address=[]
rating=[]
```

```python
#standard url and api key to use google places api, You need to request search in url format to get data from api,currently we are searching for nearest restaurants.
url="https://maps.googleapis.com/maps/api/place/textsearch/xml?"
api_key="AIzaSyDh4FXrtHWRLS4I2mtnX8X2Hw4TKbkVGu4"
#new="query=Airport&location={}&radius=10000&key={}".format(df.latlong[0],api_key)
```

```python
for i in df["latlong"]:
    new="query=restaurant&location={}&radius=1000&language=en&key={}".format(i,api_key)
    r=urllib.request.urlopen(url+new)
    soup=BeautifulSoup(r,"lxml")
    name.append(soup.find("name").text)
    lat.append(soup.find("lat").text)
    lng.append(soup.find("lng").text)
    address.append(soup.find("formatted_address").text)
```

```python
#Now we collected data into empty lists
address
```

```python
#now we will be attaching these lists to dataframe
df["Nearest Restuarant"]=name
df["lat(Restuarant)"]=lat
df["lng(Restuarant)"]=lng
df["address(Restuarant)"]=address
```
